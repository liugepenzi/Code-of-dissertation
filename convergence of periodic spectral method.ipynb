```python
import matplotlib.pyplot as plt
import numpy as np

# Read the convergence data file
data = np.loadtxt('D:/program/band-limited interpolation/convergence_data.txt')
N = data[:, 0]
error = data[:, 1]

# Generate colors cyclically
num_points = len(N)
colors = plt.cm.tab10(np.arange(num_points) % 10)

# Plotting
plt.figure(figsize=(10, 6))  # Set the figure size
for i in range(num_points):
    plt.loglog(N[i], error[i], '.', markersize=10, color=colors[i])
plt.grid(True)
plt.xlabel('N')
plt.ylabel('Error')
plt.xticks([10**i for i in range(3)])
plt.yticks([10**(-15), 10**(-10), 10**(-5), 10**(0)])
plt.title('Convergence of spectral differentiation')
plt.show()
```
