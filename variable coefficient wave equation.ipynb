```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = np.genfromtxt('wave_equation_data.csv', delimiter=',')

# Extract the x, y, and z values from the data
x = np.linspace(0, 2*np.pi, num=128)
y = np.linspace(0, 8, num=54)
z = data

# Create a 2D grid of x and y values
X, Y = np.meshgrid(x, y)

# Plot the data using a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, z)

# Set the axis ranges
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(0, 8)
ax.set_zlim(0, 5)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```
