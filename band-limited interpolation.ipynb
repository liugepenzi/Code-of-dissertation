```python
import numpy as np
import matplotlib.pyplot as plt

output_directory = "D:/program/band-limited interpolation"  # Specify the directory where the output files are saved

# Plotting parameters
plt_params = [
    {"plt": 1, "label": "plt = 1", "color": "blue"},
    {"plt": 2, "label": "plt = 2", "color": "green"},
    {"plt": 3, "label": "plt = 3", "color": "red"}
]

# Set the height of the combined figure
figure_height = 2  # Adjust this value as needed

# Create a single figure with subplots
fig, axes = plt.subplots(len(plt_params), 1, figsize=(8, len(plt_params) * figure_height))

# Plot the data
for i, params in enumerate(plt_params):
    ax = axes[i]
    
    # Read v values
    v_filename = f"{output_directory}/v_values_{params['plt']}.txt"
    x_v, v = np.loadtxt(v_filename, delimiter="\t", unpack=True)
    
    # Read p values
    p_filename = f"{output_directory}/p_values_{params['plt']}.txt"
    x_p, p = np.loadtxt(p_filename, delimiter="\t", unpack=True)
    
    # Plot v and p values
    ax.plot(x_v, v, ".", markersize=5, color=params["color"], label="v")
    ax.plot(x_p, p, linewidth=0.5, color=params["color"], label="p")
    
    ax.set_title(params["label"])
    ax.legend()

# Adjust spacing between subplots
plt.tight_layout()

# Save the combined figure
plt.savefig("combined_plots.png", dpi=300)  # Adjust the filename and dpi as needed

# Display the combined figure
plt.show()
```
