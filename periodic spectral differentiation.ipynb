```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from files
vhat = np.loadtxt("vhat.txt")
computedvhat = np.loadtxt("computedvhat.txt")
vexpSinX = np.loadtxt("vexpSinX.txt")
computedvexpSinX = np.loadtxt("computedvexpSinX.txt")

# Plotting
plt.figure(figsize=(12, 8))

# Function vhat
plt.subplot(2, 2, 1)
plt.plot(vhat[:, 0], vhat[:, 1], '.-', markersize=13)
plt.title('Function vhat')
plt.xlabel('x')
plt.ylabel('vhat')

# Spectral derivative of vhat
plt.subplot(2, 2, 2)
plt.plot(computedvhat[:, 0], computedvhat[:, 1], '.-', markersize=13)
plt.title('Spectral derivative of vhat')
plt.xlabel('x')
plt.ylabel('computedvhat')

# Function vexpSinX
plt.subplot(2, 2, 3)
plt.plot(vexpSinX[:, 0], vexpSinX[:, 1], '.-', markersize=13)
plt.title('Function vexpSinX')
plt.xlabel('x')
plt.ylabel('vexpSinX')

# Spectral derivative of exp(sin(x))
plt.subplot(2, 2, 4)
plt.plot(computedvexpSinX[:, 0], computedvexpSinX[:, 1], '.-', markersize=13)
plt.title('Spectral derivative of exp(sin(x))')
plt.xlabel('x')
plt.ylabel('computedvexpSinX')

# Adjust spacing
plt.tight_layout()

# Display the plot
plt.show()
```
