```python
import matplotlib.pyplot as plt
import numpy as np

# Load the error data from the CSV file
data = np.loadtxt("error_data.csv", delimiter=",", skiprows=1)  # Skip the header row
N = data[:, 0]  # N values from the first column
error = data[:, 1]  # Error values from the second column

plt.plot(N, error, marker='o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('N')
plt.ylabel('Error')
plt.title('Convergence of 4th-order finite differences')
plt.grid(True)

# Set custom tick labels
plt.xticks([10**i for i in range(5)], ['10^0', '10^1', '10^2', '10^3', '10^4'])
plt.yticks([10**(-15), 10**(-10), 10**(-5), 10**(0)], ['10^-15', '10^-10', '10^-5', '10^0'])

plt.show()
```
